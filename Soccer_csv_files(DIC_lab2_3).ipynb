```r
library("RSQLite")
library("sqldf")
```

```r
dbfile="/Users/vanshika/Desktop/Data Intensive Computing/database.sqlite"
sqlite = dbDriver("SQLite");
connect = dbConnect(sqlite, dbfile);
```

```r
dbListTables(connect)
```

Making the CSV file for having the Player Information.

```r
QUERY_Player<-dbGetQuery(connect, 'select p.player_api_id ,p.player_name, max(pa.overall_rating) as Overall_rating , 
max(pa.gk_positioning)+max(pa.gk_handling)+max(pa.gk_kicking)+max(pa.gk_diving)+max(pa.gk_reflexes)as GoalKeeper_Rating
from Player p JOIN Player_Attributes pa ON p.player_api_id=pa.player_api_id group by p.player_name order by 2 asc')
```

```r
QUERY_Player
```

```r
Query_Match<-dbGetQuery(connect, 'select count(*) Appearances, count(DISTINCT(m.league_id)) as Number_of_League,p.player_api_id ,p.player_name  from 
Player p JOIN Match m ON 
p.player_api_id=m.home_player_1 OR 
p.player_api_id=m.home_player_2 OR 
p.player_api_id=m.home_player_3 OR
p.player_api_id=m.home_player_4 OR
p.player_api_id=m.home_player_5 OR
p.player_api_id=m.home_player_6 OR
p.player_api_id=m.home_player_7 OR
p.player_api_id=m.home_player_8 OR
p.player_api_id=m.home_player_9 OR
p.player_api_id=m.home_player_10 OR
p.player_api_id=m.home_player_11 OR
p.player_api_id=m.away_player_1 OR
p.player_api_id=m.away_player_2 OR
p.player_api_id=m.away_player_3 OR
p.player_api_id=m.away_player_4 OR
p.player_api_id=m.away_player_5 OR
p.player_api_id=m.away_player_6 OR
p.player_api_id=m.away_player_7 OR
p.player_api_id=m.away_player_8 OR
p.player_api_id=m.away_player_9 OR
p.player_api_id=m.away_player_10 OR
p.player_api_id=m.away_player_11
group by p.player_api_id order by 2 desc')
```

```r
Query_Match
```

```r
df_merge_Player=merge(QUERY_Player, Query_Match, by.x=c("player_api_id"), by.y=c("player_api_id"))
df_merge_Player
```

```r
df_merge_Player$player_name.y<-NULL
colnames(df_merge_Player)[2]<-"Player" 
df_merge_Player
```

```r
write.csv(df_merge_Player, file = "Player.csv",row.names=FALSE)
```

Making the CSV file containing the team information.

```r
QUERY_Team<-dbGetQuery(connect, 'select
                    t.team_api_id ,t.team_long_name , MAX(IFNULL(ta.buildUpPlaySpeed,0)) as buildUpPlaySpeed,
                    MAX(IFNULL(ta.buildUpPlayDribbling,0)) as buildUpPlayDribbling,
                    MAX(IFNULL(ta.buildUpPlayPassing,0)) as buildUpPlayPassing,
                    MAX(IFNULL(ta.buildUpPlaySpeed,0))+MAX(IFNULL(ta.buildUpPlayDribbling,0))+MAX(IFNULL(ta.buildUpPlayPassing,0))as BuildUpPlay,
                    MAX(IFNULL(ta.chanceCreationPassing,0))+MAX(IFNULL(ta.chanceCreationCrossing,0))+MAX(IFNULL(ta.chanceCreationShooting,0)) as ChanceCreation,
                    MAX(IFNULL(ta.defencePressure,0))+MAX(IFNULL(ta.defenceAggression,0))+MAX(IFNULL(ta.defenceTeamWidth,0)) as Defense
                    from Team t JOIN Team_Attributes ta ON t.team_api_id=ta.team_api_id group by t.team_api_id order by 1 asc')





```

```r
QUERY_Team
```

```r
Query_Goals<-dbGetQuery(connect,'SELECT  home_team_api_id as team_api_id , home_team_goal from Match UNION ALL select away_team_api_id , away_team_goal from Match order by 1 asc')
```

```r
Query_Goals
```

```r
Goals<-sqldf('select  team_api_id, Sum(home_team_goal) as No_Of_Goals from Query_Goals group by team_api_id order by 2 desc')
```

```r
Goals
```

```r
df_merge_Team=merge(QUERY_Team, Goals, by.x=c("team_api_id"), by.y=c("team_api_id"))
df_merge_Team
```

```r
write.csv(df_merge_Team, file = "Team.csv",row.names=FALSE)
```
